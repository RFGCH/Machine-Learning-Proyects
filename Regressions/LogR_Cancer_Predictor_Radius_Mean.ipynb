Def de las funciones y variables

```python
def Sigm(x):
  return 1/1+np.exp(-x)

def Reg_Log(theta,x,y,itr,ratio):
  len = x.shape[0]
  for itrs in range(itr):
    for datos in range(len):
       for i in range(x.shape[1]):
        theta[i] -= sum(( Sigm(theta.T * x) - y)* np.reshape((x[:,i]),(len,1)))[i] * ratio/len
```

```python
def prediccion (theta,datos):
  return np.dot(datos,theta)
def ECM (Y_test,predic):
  return sum(Y_test-predic*Y_test-predic)/Y_test.shape[0]
```
