```python
import torch
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = torch.rand(20, 20)*255
plt.imshow(image, cmap='gray', vmin=0, vmax=255)
```

```python
filtro = torch.tensor([1/81]*81).reshape(9,9)
resultImage = torch.clone(image)
for i in range(image.shape[0]-8):
  for j in range(image.shape[1]-8):
    res = torch.sum(image[i:i+9,j:j+9]*filtro)
    resultImage[i+4, j+4] = res
```

```python
plt.imshow(resultImage, cmap='gray', vmin=0, vmax=255)
```

```python
def blur9by9(image):
  filtro = torch.tensor([1/81]*81).reshape(9,9)
  resultImage = torch.clone(image)
  for i in range(image.shape[0]-8):
    for j in range(image.shape[1]-8):
      res = torch.sum(image[i:i+9,j:j+9]*filtro)
      resultImage[i+4, j+4] = res
  return resultImage
```

```python
image2000 = torch.rand(2000, 2000)*255
plt.imshow(image2000, cmap='gray', vmin=0, vmax=255)
```

```python
blurred2000 = blur9by9(image2000)
print(blurred2000.shape)
plt.imshow(blurred2000, cmap='gray', vmin=0, vmax=255)
```
