```python
import torch
import numpy as np

list = ([1,2],[3,4],[5,6])
tensor1 = torch.tensor(list)
print("T1:", tensor1)

tensor2 = torch.tensor([[[1,2,3],[4,5,6]],[[1,2,3],[4,5,6]]])
print("T2:", tensor2)
```

```python
print(tensor1.size())
print(tensor2.size())
```

```python
print(tensor1[0])
print(tensor1[0,1])
```

```python
x = torch.tensor([2],dtype=torch.float,requires_grad=True)
print(x)
y = x.pow(3)
print(y)
y.backward()
x.grad
```

```python
%%time
for i in range(500):
  x = torch.randn(1000,1000)
  y = torch.randn(1000,1000)
  z = torch.matmul(x,y)
```
